```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Airport_Quarterly_Passenger_Survey.csv')
data['Date recorded'] = pd.to_datetime(data['Date recorded'])
data.sort_values(by=['Date recorded'], inplace=True)
cleaned = data.dropna()
```

```python
data.columns
```

```python
df_by_Q = pd.DataFrame(cleaned['Quarter'].unique())
for value in np.sort(cleaned['Overall satisfaction'].unique()):
    col = []
    for quarter in cleaned['Quarter'].unique():
        index = (cleaned['Quarter']==quarter) & (cleaned['Overall satisfaction']==value)
        col.append(cleaned['Overall satisfaction'][index].size / cleaned['Quarter'][cleaned['Quarter']==quarter].size)
    df_by_Q[df_by_Q.shape[1]] = col
```

```python
legend = []
p = plt.bar(df_by_Q[0],df_by_Q[1])
bottom = df_by_Q[1].copy()
legend.append(p[0])
for col in range(2,df_by_Q.shape[1]):
    p = plt.bar(df_by_Q[0],df_by_Q[col],bottom=bottom)
    bottom += df_by_Q[col]
    legend.append(p[0])
plt.xticks(rotation=45)
plt.xlabel('Quarter')
plt.ylabel(data.columns[-1])
plt.legend(legend, ('0','1','2','3','4','5'),loc='best', bbox_to_anchor=(1, 0.8))
plt.show()
```

```python
var = 3
df_by_Q = pd.DataFrame(cleaned['Quarter'].unique())
for value in np.sort(cleaned.iloc[:,var].unique()):
    column = []
    for quarter in cleaned['Quarter'].unique():
        index = (cleaned['Quarter']==quarter) & (cleaned.iloc[:,var]==value)
        column.append(cleaned.iloc[:,var][index].size / cleaned['Quarter'][cleaned['Quarter']==quarter].size)
    df_by_Q[df_by_Q.shape[1]] = column
    
#plot stacked column chart
legend = []
p = plt.bar(df_by_Q[0],df_by_Q[1])
bottom = df_by_Q[1].copy()
legend.append(p[0])
for col in range(2,df_by_Q.shape[1]):
    p = plt.bar(df_by_Q[0],df_by_Q[col],bottom=bottom)
    bottom += df_by_Q[col]
    legend.append(p[0])
plt.xticks(rotation=45)
plt.xlabel('Quarter')
plt.ylabel(data.columns[var])
plt.legend(legend, ('0','1','2','3','4','5'),loc='best', bbox_to_anchor=(1, 0.8))
plt.show()
```
